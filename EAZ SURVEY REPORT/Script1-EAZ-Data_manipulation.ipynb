```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(pwd + "/EAZ members Survey Responses.xlsx")
dataset 
```

```python
dataset_modified = dataset.copy()
dataset_modified
```

```python
id_vars = list(dataset_modified.columns)[0 : 1]
value_vars = list(dataset_modified.columns)[1 : ]
value_vars
```

```python
dataset_melted = dataset_modified.melt(id_vars= id_vars,value_vars= value_vars,var_name= "Question", value_name="Answer")
dataset_melted
```

```python
# Filter responses for the Second question
question_2_responses = dataset_melted[dataset_melted['Question'] == 'What is Your Sex ']

# Calculate value counts for each answer
sex_counts = question_2_responses['Answer'].value_counts()

# Visualize the distribution of responses
plt.figure(figsize=(8, 4))
sex_counts.plot(kind='bar', color='green')
plt.title('Distribution of Respondents by Field of Work')
plt.xlabel('Field of Work')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Filter responses for the question about sector
sector_responses = dataset_melted[dataset_melted['Question'] == 'Which sector are you working in']

# Calculate value counts for each sector
sector_counts = sector_responses['Answer'].value_counts()

# Keep only the top 5 most selected sectors
top_five_sectors = sector_counts.head()

# Calculate the count for "Other"
other_sector_count = sector_counts.sum() - top_five_sectors.sum()

# Create a DataFrame with the top 5 sectors
top_five_sectors_with_other = pd.DataFrame({'Sector': top_five_sectors.index, 'Count': top_five_sectors.values})

# Add a row for "Other"
other_sector_row = pd.DataFrame({'Sector': ['Other'], 'Count': [other_sector_count]})
top_five_sectors_with_other = pd.concat([top_five_sectors_with_other, other_sector_row], ignore_index=True)

# Visualize the distribution of responses
plt.figure(figsize=(8, 4))
sector_plot = sns.barplot(x='Sector', y='Count', data=top_five_sectors_with_other, color='green')

# Add counts on top of the bars
for index, row in top_five_sectors_with_other.iterrows():
    sector_plot.text(index, row['Count'] + 0.05, row['Count'], color='black', ha="center")

plt.title('Distribution of Participants by Sector')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Calculate the total count of respondents
total_count = top_five_sectors_with_other['Count'].sum()

# Calculate the percentage for each sector
top_five_sectors_with_other['Percentage'] = (top_five_sectors_with_other['Count'] / total_count) * 100

# Visualize the distribution of responses
plt.figure(figsize=(8, 4))
sector_plot = sns.barplot(x='Sector', y='Count', data=top_five_sectors_with_other, color='Green')

# Add percentages on top of the bars
for index, row in top_five_sectors_with_other.iterrows():
    sector_plot.text(index, row['Count'] + 0.05, f"{row['Percentage']:.2f}%", color='black', ha="center")

plt.title('Distribution of Participants by Sector')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter responses for the question about level of experience
experience_responses = dataset_melted[dataset_melted['Question'] == 'What is your level of experience?']

# Calculate value counts for each level of experience
experience_counts = experience_responses['Answer'].value_counts()

# Keep only the top 5 most selected levels of experience
top_five_experience = experience_counts.head(5)

# Calculate the total count of respondents
total_count = top_five_experience.sum()

# Calculate the percentage for each level of experience
top_five_experience_percentage = (top_five_experience / total_count) * 100

# Visualize the distribution of responses
plt.figure(figsize=(8, 4))
experience_plot = sns.barplot(x=top_five_experience_percentage.index, y=top_five_experience_percentage.values, color='orange')

# Add percentages on top of the bars
for index, value in enumerate(top_five_experience_percentage):
    experience_plot.text(index, value + 0.5, f"{value:.2f}%", color='black', ha="center")

plt.title('Distribution of Respondents by Level of Experience')
plt.xlabel('Level of Experience')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter responses for the question about paid-up membership
paid_up_responses = dataset_melted[dataset_melted['Question'] == 'Are you a Paid Up Member of The Economics Association of Zambia (EAZ)?']

# Calculate value counts for each response
paid_up_counts = paid_up_responses['Answer'].value_counts()

# Visualize the distribution of responses
plt.figure(figsize=(8, 4))
sns.barplot(x=paid_up_counts.index, y=paid_up_counts.values, color='darkgreen')

plt.title('Distribution of Paid Up Members')
plt.xlabel('Paid Up Membership Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
